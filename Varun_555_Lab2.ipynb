```python
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM
import torch
```

```python
gpt2_tokenizer = AutoTokenizer.from_pretrained("gpt2")
gpt2_model = AutoModelForCausalLM.from_pretrained("gpt2")

bart_tokenizer = AutoTokenizer.from_pretrained("facebook/bart-base")
bart_model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-base")

flan_tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
flan_model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")
```

```python
prompts = [
    "A lonely robot discovers a hidden message from the past.",
    "In a city where dreams are bought and sold, a boy tries to steal one.",
    "An old woman finds a book that tells the future—and it shows her death."
]
```

```python
def generate_story(prompt, model, tokenizer, max_len=200):
    inputs = tokenizer(prompt, return_tensors="pt")
    if isinstance(model, AutoModelForCausalLM):
        output = model.generate(
            **inputs,
            max_length=max_len,
            temperature=0.8,
            top_k=50,
            top_p=0.95,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )
    else:
        output = model.generate(
            **inputs,
            max_length=max_len,
            temperature=0.8,
            top_k=50,
            top_p=0.95,
            do_sample=True
        )
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
for prompt in prompts:
    print(f"\n🔹 Prompt: {prompt}\n")

    print("📘 GPT-2 Story:")
    print(generate_story(prompt, gpt2_model, gpt2_tokenizer))

    print("\n📙 BART Story:")
    print(generate_story(prompt, bart_model, bart_tokenizer))

    print("\n📗 FLAN-T5 Story:")
    print(generate_story(prompt, flan_model, flan_tokenizer))
```

```python
# 📌 Install nltk + rouge_score
!pip install nltk rouge-score --quiet
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer

nltk.download('punkt')
nltk.download('punkt_tab')
smoothie = SmoothingFunction().method4
rouge = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)

# ✅ Reference story for testing
reference = "The robot found an old recording hidden under debris. It was a message from its creator, explaining its purpose. The robot now had meaning."

# ✅ Evaluate model outputs for 1st prompt
prompt = prompts[0]
gpt2_story = generate_story(prompt, gpt2_model, gpt2_tokenizer)
bart_story = generate_story(prompt, bart_model, bart_tokenizer)
flan_story = generate_story(prompt, flan_model, flan_tokenizer)

# Tokenize reference and predictions
ref_tokens = nltk.word_tokenize(reference.lower())
predictions = {
    "GPT-2": gpt2_story,
    "BART": bart_story,
    "FLAN-T5": flan_story
}

# ✅ Print BLEU & ROUGE scores
for model_name, story in predictions.items():
    pred_tokens = nltk.word_tokenize(story.lower())
    bleu = sentence_bleu([ref_tokens], pred_tokens, smoothing_function=smoothie)
    rouge_scores = rouge.score(reference, story)

    print(f"\n📊 Evaluation for {model_name}:")
    print(f"🟦 BLEU Score: {bleu:.4f}")
    print(f"🟥 ROUGE-1: {rouge_scores['rouge1'].fmeasure:.4f}")
    print(f"🟩 ROUGE-L: {rouge_scores['rougeL'].fmeasure:.4f}")
```

```python
# 📌 Install Gradio
!pip install gradio --quiet

import gradio as gr

# ✅ Function for Gradio
def generate_flan_story(prompt, temperature=0.8, max_len=300):
    inputs = flan_tokenizer(prompt, return_tensors="pt")
    output = flan_model.generate(
        **inputs,
        max_length=max_len,
        temperature=temperature,
        top_k=50,
        top_p=0.95,
        do_sample=True
    )
    return flan_tokenizer.decode(output[0], skip_special_tokens=True)

# ✅ Gradio Interface
gr.Interface(
    fn=generate_flan_story,
    inputs=[
        gr.Textbox(label="Enter your story prompt here"),
        gr.Slider(minimum=0.1, maximum=1.5, value=0.8, label="Temperature"),
        gr.Slider(minimum=100, maximum=500, value=300, label="Max Length")
    ],
    outputs=gr.Textbox(label="Generated Story"),
    title="📝 FLAN-T5 Story Generator",
    description="Enter a creative prompt and generate a story using FLAN-T5!"
).launch()
```

| **Aspect**             | **GPT-2**                                   | **BART**                        | **FLAN-T5**                          |
| ---------------------- | ------------------------------------------- | ------------------------------- | ------------------------------------ |
| **Architecture**       | Decoder-only (Causal)                       | Encoder-Decoder                 | Encoder-Decoder (Instruction-tuned)  |
| **Training Objective** | Predict next token (left-to-right)          | Denoising autoencoder (Seq2Seq) | Instruction-based generation         |
| **Output Style**       | Imaginative but inconsistent and repetitive | Very concise and literal        | Short, vague, but logical            |
| **Story Quality**      | Sometimes derails into unrelated loops      | Mostly repeats the prompt only  | Slightly better attempt at structure |
| **Creativity**         | 🔸 Medium-High                              | 🔸 Low                          | 🔸 Medium                            |
| **Coherence**          | ⚠️ Weak                                     | ✅ Strong                        | ⚠️ Moderate                          |
| **BLEU Score**         | 0.0065                                      | 0.0151 ✅                        | 0.0133                               |
| **ROUGE-1**            | 0.0860                                      | 0.3529 ✅                        | 0.2927                               |
| **ROUGE-L**            | 0.0538                                      | 0.2941 ✅                        | 0.1463                               |
| **Best Use Case**      | Open-ended story generation                 | Summarization or paraphrasing   | Instruction-following & QA           |
| **Limitation**         | Repetition & hallucination                  | Lacks creativity                | Very brief output                    |



Among the three models, BART performed best in coherence and fluency as reflected by the highest BLEU and ROUGE scores, even though it lacked creativity. GPT-2 was the most imaginative but often generated erratic or repetitive stories. FLAN-T5 was concise and sometimes logical, but failed to generate detailed content beyond a sentence or two.

Justification:
Balanced performance in coherence and evaluation scores
Better control over structure due to its encoder-decoder format
Ideal for short-form, logical, and grammatically correct outputs
